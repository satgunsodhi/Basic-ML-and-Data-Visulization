```python
#importing required packages
import numpy as np
import pandas as pd
from random import choice
import matplotlib.pyplot as plt
```

```python
#loading the dataset
dataset = pd.read_csv('C:\\Users\\Satgu\\projects\\mozilla\\Linear Regression Model\\Housing.csv').to_numpy()
```

```python
#initialising the train and test arrays
x_train = []
y_train = []
x_test = []
y_test = []
```

```python
#making a split between the two from the dataset
for i in range(0,len(dataset)):
    switch = choice([1,2])
    if switch == 1:
        x_train.append(dataset[i][1])
        y_train.append(dataset[i][0])
    else:
        x_test.append(dataset[i][1])
        y_test.append(dataset[i][0])
```

```python
# converting the lists to numpy arrays
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
print((x_train))
```

```python
#training the model based on mean squared error
n = len(x_train)
alpha = 0.00000000001

a_0 = 0
a_1 = 0

epochs = 1000

for i in range(0,n):
    y = a_0 + a_1 * x_train
    error = y - y_train
    mean_sq_er = np.sum(error**2)
    mean_sq_er = mean_sq_er/n
    a_0 = a_0 - alpha * 2 * np.sum(error)/n
    a_1 = a_1 - alpha * 2 * np.sum(error * x_train)/n
    print(mean_sq_er)
```

```python
#plotting the required graph against test data
y_plot = a_0 + a_1*x_test
plt.figure(figsize=(10,10))
plt.scatter(x_test,y_test,color='red',label='GT')
plt.plot(x_test,y_plot,color='black',label = 'pred')
plt.legend()
plt.show()
```
